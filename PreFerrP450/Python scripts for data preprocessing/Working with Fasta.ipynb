# Creates a table with all information of fasta and species name

```python
import Bio
import re
import pandas as pd
from Bio import SeqIO
from Bio.Seq import Seq
from Bio import Entrez

filename_input="output/sirinsp450permutationsof4/listfalsepositive.fasta"
filename_output=filename_input+"info.csv"
tableofproteins=pd.DataFrame(columns=['NCBI Reference','Gene locus tag','Position','Direction','Species','Strain',"Definition"])
Entrez.email = "friederike.biermann@uni-jena.de"  # Always tell NCBI who you are
listfasta=[]
liststrains=[]
def find_between( s, first, last ):
    try:
        start = s.index( first ) + len( first )
        end = s.index( last, start )
        return s[start:end]
    except ValueError:
        return ""
def find_infrontof( s, last ):
    try:
        start = 0
        end = s.index( last, start )
        return s[start:end]
    except ValueError:
        return ""

for record in SeqIO.parse(filename_input, "fasta"):
    
    Direction=find_between(record.id,"(",")")
    Position=find_between(record.id,"[","]")
    locustag=find_infrontof(record.id,"|")
    accessionnumber = find_between(record.id,"|","|")
    handle = Entrez.efetch(db="nucleotide", id=accessionnumber, rettype="gb", retmode="xml")
    result = Entrez.read(handle)
    name = result[0]["GBSeq_source"]
    definition=str(result[0]["GBSeq_definition"])
    if name in liststrains:
        print("strain already existing: ",name)
    else: listfasta.append(record)
    liststrains.append(name)
    
    if "strain" in definition:
        match = re.search("strain (\w+)", definition)
        strainnumber= match.group()
        strain= name+" "+strainnumber
    else:
        strain= name
    handle.close()
    new_row={'NCBI Reference':accessionnumber,'Gene locus tag':locustag,'Position':Position,'Direction':Direction,'Species':name,'Strain':strain,"Definition":definition}
    tableofproteins = tableofproteins.append(new_row, ignore_index=True)
    
SeqIO.write(listfasta, "newlistfasta.fasta", "fasta")
tableofproteins.to_csv(filename_output, index=False) 
```
