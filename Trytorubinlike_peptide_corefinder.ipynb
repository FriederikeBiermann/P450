# Find reading frames in genomes next to P450s

```python
import Bio
import re
import pandas as pd
from Bio import SeqIO
from Bio.Seq import Seq
from Bio import Entrez

#tableofproteins.to_csv(filename_output, index=False)

front=2000
back=2000

stopcodon1=('TGA', 'TAA', 'TAG')
startcodon1=('ATG',"GTG")
stopcodon2=("TCA","TTA","CTA")
startcodon2=("CAT","CAC")

#insert length of protein in aa
minlengthreadingframe=20
maxlengthreadingframe=40

filename_input="Output/sirinsp450permutationsof4/p450ncbiallpreoteins.fastalistpositive.fasta"
filename_output=filename_input+"infonew.csv"
tableofproteins=pd.DataFrame(columns=['Protein Reference','NCBI Reference','Species',"Definition", "gb_whole_genome_accession","gb_whole_genome_position","seqencednafragment","openreadingframes","coreproteins"])
Entrez.email = "friederike@biermann-erfurt.de"
filenamereadingframes="Output/sirinsp450permutationsof4/readingframes.txt"
listfasta=[]
liststrains=[]

def find_between( s, first, last ):
    try:
        start = s.index( first ) + len( first )
        end = s.index( last, start )
        return s[start:end]
    except ValueError:
        return ""
def find_infrontof( s, last ):
    try:
        start = 0
        end = s.index( last, start )
        return s[start:end]
    except ValueError:
        return ""
def find_after( s, first ):
    try:
        start =s.index( first ) + len( first )
        end = len(s)
        return s[start:end]
    except ValueError:
        return ""
    
listreadingframe=[]
with open(filenamereadingframes, 'r') as f:
    listreadingframe = [line.rstrip('\n') for line in f]
    
#searching for reading frames in fastas
for record in SeqIO.parse(filename_input, "fasta"):
    try:
        rid=record.id
        print (rid)
        if "ref" in rid:
            #find id in fasta header
            gi = find_between(record.id,"gi|","|")
            ref= find_between(record.id,"ref|","|")
            # find all nucleotides this reference protein is part of
            handle = Entrez.elink(dbFrom = "protein", db = "nucleotide",id=gi)
            record = str(Entrez.read(handle))
            handle.close()
            listgenes=re.findall("(?<=Id': ')[0-9]+",record)
            listgenes=listgenes[:int(len(listgenes)/2)]
            for name1 in listgenes:
                # get sequence of every genome
                dnahandle1 = Entrez.efetch(db="nucleotide", id=name1, rettype='gbwithparts',retmode="xml")
                record = Entrez.read(dnahandle1)
                dnahandle1.close()
                definition=str(record[0]["GBSeq_definition"])
                species = record[0]["GBSeq_organism"]
                for match in re.finditer(ref, str(record)):
                    #find p450 in sequence/ in genbank file and the proteins before and after
                    proteinspan1=match.span()[0]
                   # print (proteinspan1)
                    proteinspan2=int((proteinspan1))-1000
                    proteinspan3=int((proteinspan1))-3000
                    proteinspan4=int((match.span()[1]))+1000
                    startdna=re.search("(?<='GBInterval_from': ')[0-9]+", str(record)[proteinspan2:proteinspan1])
                    enddna=re.search("(?<='GBInterval_to': ')[0-9]+", str(record)[proteinspan2:proteinspan1])                    
                    startdna2=re.search("(?<='GBInterval_from': ')[0-9]+", str(record)[match.span()[1]:proteinspan4])                 
                    enddna2=re.search("(?<='GBInterval_to': ')[0-9]+", str(record)[match.span()[1]:proteinspan4])                    
                    startdna3=re.search("(?<='GBInterval_from': ')[0-9]+", str(record)[proteinspan3:proteinspan2])
                    enddna3=re.search("(?<='GBInterval_to': ')[0-9]+", str(record)[proteinspan3:proteinspan2])
                    p450count=0
                    listp450=[]
                    
                    #find exact p450
                    for match in re.finditer("P450", str(record)):
                        p450count=p450count+1
                        proteinspan1=match.span()[0]+500
                        proteinspan2=int((proteinspan1))-1000
                        startp450=re.search("(?<='GBInterval_from': ')[0-9]+", str(record)[proteinspan2:proteinspan1])  
                        endp450=re.search("(?<='GBInterval_to': ')[0-9]+", str(record)[proteinspan2:proteinspan1])                        
                        startp450=startp450.group(0)
                        endp450=endp450.group(0)
                        string= str(startp450)+"-"+str(endp450)
                        listp450.append(string)
                    startdna=startdna.group(0)
                    enddna=enddna.group(0)
                    startdna2=startdna2.group(0)
                    enddna2=enddna2.group(0)
                    startdna3=startdna3.group(0)
                    enddna3=enddna3.group(0)
                if len(enddna)>0 and len(startdna)>0:
                    framebegin=int(startdna)-front
                    frameend=int(enddna)+back
                    #get sequence of whole genomic background beore/after protein
                    dnahandle = Entrez.efetch(db="nucleotide", id=name1, rettype="fasta",retmode="txt")
                    for seq_record in SeqIO.parse(dnahandle, "fasta"):
                        complete_sequence=seq_record.seq
                        chosensequence=str(complete_sequence[framebegin:frameend])
                    dnahandle.close()
                    listopenframe=""
                    listcore=""
                    for frame in listreadingframe:
                        #actual corefinder- searches for open reading frames near protein
                        for match in re.finditer(frame,chosensequence):
                           # print(len(match.group()))
                            begin=match.span()[0]+framebegin
                            end=match.span()[1]+framebegin
                            # make sure core is not in things annotated in genbank file
                            if int(startdna)<begin<int(enddna)or int(startdna)<end<int(enddna)or int(enddna)<begin<int(startdna)or int(enddna)<end<int(startdna) :
                                   pass
                            if int(startdna2)<begin<int(enddna2)or int(startdna2)<end<int(enddna2)or int(enddna2)<begin<int(startdna2)or int(enddna2)<end<int(startdna2) :
                                   pass
                            if int(startdna3)<begin<int(enddna3)or int(startdna3)<end<int(enddna3)or int(enddna3)<begin<int(startdna3)or int(enddna3)<end<int(startdna3) :
                                   pass
                            else:                               
                                if match.group()[:3] in startcodon1:
                                    # if forward strand
                                    coredna=Seq(match.group()[-21:])
                                    aa=Seq(match.group()).translate()
                                    coreaa=coredna.translate()
                                     #tests if stop codon included
                                    if "*" in aa[:-1]:
                                         pass
                                    if "*" not in aa[:-1] and "*" not in coreaa[:-1]:
                                        print ("+",coreaa)
                                        if str(end) and str(begin) not in listcore:
                                            listopenframe=listopenframe+", "+match.group()+", "+str(begin)+", "+str(end)+"(+)"
                                            listcore=listcore+", "+str(coreaa)+", "+str(begin)+", "+str(end)+"(+)"
                                if match.group()[:3] in stopcodon2:
                                    #if reverse strand
                                    reversedna=Seq(match.group()).back_transcribe().reverse_complement()
                                    reversedna=str(reversedna)
                                    aa=Seq(reversedna).translate()
                                    coredna=reversedna[-21:]
                                    coreaa=Seq(coredna).translate()
                                    #tests if stop codon included
                                    if "*" in aa[:-1]:
                                         pass
                                    if "*" not in aa[:-1] and "*" not in coreaa[:-1]:
                                        if str(end) and str(begin) not in listcore:
                                            print ("-",coreaa)
                                            listopenframe=listopenframe+", "+str(reversedna)+", "+str(begin)+", "+str(end)+"(-)"
                                            listcore=listcore+", "+str(coreaa)+", "+str(begin)+", "+str(end)+"(-)"
                    position4=str(startdna)+".."+str(enddna)
                    # add new row with information to table
                    new_row={'Protein Reference':ref,'NCBI Reference':name1,'Species':species,"Definition":definition,"gb_whole_genome_accession":name1,"gb_whole_genome_position":position4,"seqencednafragment":chosensequence,"openreadingframes":str(listopenframe),"coreproteins":str(listcore)}
                    
                    tableofproteins = tableofproteins.append(new_row, ignore_index=True)
        
        if "gb" in rid:
            # same thing but not with refernce protein but with exact protein
            gi = find_between(record.id,"gi|","|")
            ref= find_between(record.id,"gb|","|")
            handle = Entrez.efetch(db="protein", id=gi, rettype="gb", retmode="xml")
            records = Entrez.parse(handle)            
            for record in records:
                definition=record["GBSeq_definition"]
                name = record["GBSeq_organism"]
                name1=find_after(name,"n ")
                position = str(record['GBSeq_feature-table'])
                position2=find_between(position, "locus_tag","transl_table")
                position3=find_between(position2, "{'GBQualifier_name': 'coded_by', 'GBQualifier_value': '","'}, {'GBQualifier_name': 'note'") 
                position4=find_after(position3,":")[0:25]
                species = record["GBSeq_organism"]
                #get dna of strain
                startdna=find_infrontof( position4, ".")
                enddna=find_after(position4, "..")
                startdna= re.findall(r'\d', startdna)
                startdna="".join(startdna)
                enddna= re.findall(r'\d', enddna)
                enddna="".join(enddna)
                if len(enddna)>0 and len(startdna)>0:
                    framebegin=int(startdna)-front
                    frameend=int(enddna)+back
                    dnahandle = Entrez.efetch(db="nucleotide", id=name1, rettype="fasta",retmode="txt")
                    chosensequence=""
                    for seq_record in SeqIO.parse(dnahandle, "fasta"):
                        complete_sequence=seq_record.seq
                        chosensequence=str(complete_sequence[framebegin:frameend])
                    dnahandle.close()
                    listopenframe=""
                    listcore=""
                    for frame in listreadingframe:
                        for match in re.finditer(frame,chosensequence):            
                            begin=match.span()[0]+framebegin
                            end=match.span()[1]+framebegin
                            if int(startdna)<begin<int(enddna)or int(startdna)<end<int(enddna)or int(enddna)<begin<int(startdna)or int(enddna)<end<int(startdna) :
                                   pass
                            else:
                                if match.group()[:3] =="ATG":
                                    aa=Seq(match.group()).translate()
                                    coreaa=aa[-7:]
                                    if "*" in aa[:-1]:
                                         pass
                                    if "*" not in aa[:-1] and "*" not in coreaa[:-1]:
                                        if str(end) and str(begin) not in listcore:
                                            listopenframe=listopenframe+", "+match.group()+", "+str(begin)+", "+str(end)+"(+)"
                                            listcore=listcore+", "+str(coreaa)+", "+str(begin)+", "+str(end)+"(+)"
                                if match.group()[:3] in stopcodon2:
                                    reversedna=Seq(match.group()).back_transcribe().reverse_complement()
                                    reversedna=str(reversedna)
                                    aa=Seq(reversedna).translate()
                                    coreaa=aa[-7:]
                                    if "*" in aa[:-1]:
                                         pass
                                    if "*" not in aa[:-1] and "*" not in coreaa[:-1]:
                                        if str(end) and str(begin) not in listcore:
                                            print (coreaa)
                                            listopenframe=listopenframe+", "+str(reversedna)+", "+str(begin)+", "+str(end)+"(-)"
                                            listcore=listcore+", "+str(coreaa)+", "+str(begin)+", "+str(end)+"(-)"
                    position4=str(startdna)+':'+str(enddna)
                    new_row={'Protein Reference':ref,'NCBI Reference':name1,'Species':species,"Definition":definition,"gb_whole_genome_accession":name1,"gb_whole_genome_position":position4,"seqencednafragment":chosensequence,"openreadingframes":str(listopenframe),"coreproteins":str(listcore)}
                    tableofproteins = tableofproteins.append(new_row, ignore_index=True)
                    handle.close()
    except: pass
tableofproteins.to_csv(filename_output, index=False) 
```
